This tutorial aims at showing different source of colors that can be used to colour any plot used in the book. Note that this notebook is to facilitate the setting of the argument *cmap* but do refer to the matplotlib webseite for further references: https://matplotlib.org/examples/color/colormaps_reference.html.<br>
Note that we will take into account the *size* of the ball as well: in a real-world dataset, this might refer to the magnitude of another feature that might take into account in the evaluation of the phenomenon.

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
def generate_scatter(N,cmap):
    sample = N
    x = np.random.rand(sample)
    y = np.random.rand(sample)
    classes = np.random.rand(sample)
    area = (30 * np.random.rand(sample))**2 #size of the ball
    plt.scatter(x, y, s=area, c=classes, alpha=0.5,cmap=cmap)
    plt.show()
```

```python
generate_scatter(100,cmap='winter')
```

```python
generate_scatter(100,cmap='autumn')
```
